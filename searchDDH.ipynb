```python
import json
import requests


indicators = {}

for url in urls:
    print(f"Processing {url}")
    response = requests.get(url.strip())
    data = json.loads(response.text)

    for concept in data["source"][0]["concept"]:
        print(f"Found {len(concept['variable'])}")
        for variable in concept["variable"]:
            name = variable["name"]
            if name in indicators:
                indicators[name] += 1
            else:
                indicators[name] = 1

for name, count in indicators.items():
    print(f"{name}: {count}")
```
